# Visualize dataset using PCA from scratch

```python
label = l.head(20000)
data = d.head(20000)
print(data.shape)
print(label.shape)
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardized_data = StandardScaler().fit_transform(data)
```

```python
print(standardized_data)
```

```python
sample_data = standardized_data
covar_matrix = np.matmul(sample_data.T,sample_data)
print(covar_matrix.shape)
```

```python
from scipy.linalg import eigh
values,vector = eigh(covar_matrix,eigvals=(782,783))
```

```python
print(vector)
```

```python
print(vector.shape)
vectors = vector.T
```

```python
new_coordinates = np.matmul(vectors , sample_data.T)
```

```python
new_coordinates
```

```python
new_coordinates = np.vstack((new_coordinates,label)).T
```

```python
df = pd.DataFrame(data = new_coordinates, columns=("1st Principal Component","2nd Principal Component","label"))
```

```python
df
```

```python
sns.FacetGrid(df,hue='label',size=6).map(plt.scatter,'1st Principal Component','2nd Principal Component').add_legend()
```

# MNIST PCA using Scikit-learn

```python
from sklearn import decomposition
pca=decomposition.PCA()
```

```python
pca.n_components = 2
```

```python
label = l.head(7000)
data = d.head(7000)
```

```python
standardized_data = StandardScaler().fit_transform(data)
```

```python
pca_data = pca.fit_transform(standardized_data)
```

```python
print(pca_data.shape)
```

```python
pca_data = np.vstack((pca_data.T , label)).T
pca_df = pd.DataFrame(data= pca_data,columns = ("1","2","Label"))
sns.FacetGrid(pca_df,hue='Label',size=6).map(plt.scatter,"1","2").add_legend()
```

# MNIST using t-SNE

```python
from sklearn.manifold import TSNE
```

```python
data = standardized_data[0:1000,:]
labels = label[0:1000]
model = TSNE(n_components = 2,random_state=0)
```

```python
tsne_data = model.fit_transform(data)
tsne_data = np.vstack((tsne_data.T,labels)).T
tsne_df = pd.DataFrame(data=tsne_data,columns=('Dim1','Dim2','label'))
```

```python
sns.FacetGrid(tsne_df,hue='label',size=6).map(plt.scatter,'Dim1','Dim2').add_legend()
```

```python
model = TSNE(n_components = 2,random_state=0,perplexity=10)
tsne_data = model.fit_transform(data)
tsne_data = np.vstack((tsne_data.T,labels)).T
tsne_df = pd.DataFrame(data=tsne_data,columns=('Dim1','Dim2','label'))
sns.FacetGrid(tsne_df,hue='label',size=6).map(plt.scatter,'Dim1','Dim2').add_legend()
```
